```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#manually in numpy

#The list of numbers:
z = [1,2,3]

#Compute softmax result:
num = np.exp(z)
print(num)

den = np.sum(np.exp(z))

sigma = num/den

print(sigma)
print(np.sum(sigma))
```

```python
#repeat w/ some random integers:
z = np.random.randint(-5, high=15, size=25)
print(z)
```

```python
#compute softmax result
num = np.exp(z)
den = np.sum(num)
sigma = num / den

#compare
plt.plot(z,sigma,'ko')
plt.xlabel('Original Number (z)')
plt.ylabel('Softmax $\sigma$')

plt.yscale('log')
plt.title('$\sum\sigma$ = %g' %np.sum(sigma))
plt.show()
```
